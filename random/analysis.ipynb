```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import jupytertools
from pylab import rcParams
rcParams['figure.figsize'] = 8, 6
matplotlib.rc('text', usetex=True)
matplotlib.rc('font', family='serif')
matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20)
matplotlib.rcParams['xtick.major.pad'] = 8
matplotlib.rcParams['ytick.major.pad'] = 8
```

## Chi-squared

```python
bin_values = np.genfromtxt("trimmed_output", delimiter=" ")
```

```python
x = bin_values[:,0]
y = bin_values[:,1]
```

```python
jupytertools.plotting.darkPlot(x,y,"r.")
```

```python
fit = np.polyfit(x,y,1)
fit_fn = np.poly1d(fit) 

plt.plot(x,y, 'g.')
plt.plot(x,fit_fn(x),"-k")
plt.savefig("img/chi-squared-dist",dpi=300,bbox_inches='tight')
plt.show()
```

```python
print("mean:", np.mean(y))
```

## Frequrency

```python
frequencies = np.genfromtxt("trimmed_frequency",delimiter=" ")

x = frequencies[:,0]
y1 = frequencies[:,1]
y2 = frequencies[:,2]
fit1 = np.polyfit(x,y1,1)
fit_fn1 = np.poly1d(fit1)
fit2 = np.polyfit(x,y2,1)
fit_fn2 = np.poly1d(fit2)
print("mean 1",np.mean(y1))
print("mean 0",np.mean(y2))
```

```python
plt.plot(x,y1,'g.')
plt.plot(x,fit_fn1(x), '-k')
plt.savefig("img/frequency-1",dpi=300,bbox_inches='tight')
plt.show()
plt.plot(x,y2,'r.')
plt.plot(x,fit_fn2(x), '-k')
plt.savefig("img/frequency-0",dpi=300,bbox_inches='tight')
plt.show()
```
